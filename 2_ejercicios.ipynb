5. Dado el arreglo `a = np.array([1, np.nan, 3])`, determina si contiene valores NaN.

7. Crea un arreglo que contenga los valores 1, 7, 13 y 105, y determina la cantidad de memoria que ocupa el arreglo.
8. Crea un arreglo de 10 ceros; uno de 10 unos; y uno de 10 cincos.

```python
import numpy as np
```

```python
# ejercicio 1#
a = np.array([1, 2, 3, 4])
print(np.all(a != 0))
```

```python
# ejercicio 2#
a = np.array([0, 0, 2, 0])
print(np.any(a != 0))
```

```python
# ejercicio 3#

a = np.array([1, 2, np.nan, np.inf])
comprobacion = np.any(np.isnan(a) | np.isinf(a))
print(comprobacion)
```

```python
# ejercicio 4#
a = np.array([1, 2, np.inf])
comprobacion = np.any(np.isinf(a))
print(comprobacion)
```

```python
# ejercicio 5#
a = np.array([1, np.nan, 3])
comprobacion = np.any(np.isnan(a))
print(comprobacion)
```

```python
# ejercicio 6#
a = np.array([1, 3, 5])
b = np.array([2, 3, 4])
resultado = np.array([a > b, a >= b, a < b, a <= b])
print(resultado)
```

```python
# ejercicio 7#
a = np.array([1, 7, 13, 105])
print(a.nbytes)
```

```python
# ejercicio 8#
a = np.zeros(10)
b = np.ones(10)
c = np.full(10, 5)

print(a)
print(b)
print(c)
```
